```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
import warnings 
warnings.filterwarnings("ignore") # use to igonre the depricate function 
```

```python
df=pd.read_csv(r"C:\Users\Dell\Desktop\IPL.csv")
df
```

```python
# Basic Information
df.info()
```

```python
# Check the size of rows and columns of the dataset
print(f"your row is {df.shape[0]} and your columns are {df.shape[1]}")
```

```python
# Now let's see how many columns have null values in total.
df.isnull().sum()
```

```python
# Which team won the most matches?
matches_wins=df['match_winner'].value_counts()
matches_wins
sns.barplot(x=matches_wins.values,y=matches_wins.index,palette='rainbow')
```

```python
# Toss Decision Trends
# toss=df['toss_decision'].value_counts()
plt.title("Toss Decision Trends")
sns.countplot(x=df['toss_decision'])
```

```python
# Toss Winner vs Match Winner
count=df[df['toss_winner']==df['match_winner']]['match_id'].count()
percent=(count*100)/df.shape[0]
percent.round(2)
```

```python
# How do teams win? (Runs vs Wickets)
sns.countplot(x=df['won_by'])
```

```python
# key player performance 
# most "Player of the match" Awards 
count=df['player_of_the_match'].value_counts().head(10)
count
```

```python
sns.barplot(x=count.values,y=count.index,palette='mako')
plt.title("top 10 player with man of the match")
```

```python
# 5 Top scores 
highest=df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head()
highest.plot(kind='barh')
```

```python
# 10 bowling figures
df['highest_wickets']=df['best_bowling_figure'].apply(lambda x:x.split('--')[0])
df['highest_wickets']=df['highest_wickets'].astype(int)
top_bowlers=df.groupby(by='best_bowling')['highest_wickets'].sum().sort_values(ascending=False).head(10)
top_bowlers.plot(kind='barh')
```

```python
# Venue Analysis
# Most Matches Played by Venue
venue_count=df['venue'].value_counts()
sns.barplot(y=venue_count.index,x=venue_count.values,palette='PiYG')
```

```python
# Who won the highest margin by runs?
df[df['won_by']=='Runs'].sort_values(by='margin',ascending=False).head(1)[['match_winner','match_winner']]
```

```python
# Which player had the highest individual score?
df[df['highscore']==df['highscore'].max()][['top_scorer','highscore']]
```

```python
# Which bowler had the best bowling figures?
bestBowler=df.sort_values(by='highest_wickets',ascending=False).head()
bestBowler[['best_bowling','best_bowling_figure']]
```
